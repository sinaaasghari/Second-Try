```python
from datasets import load_dataset
from huggingface_hub import login
login('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')

Test_Data=load_dataset('OmidAghili/Image_Classification')
```

```python
x_test,y_test=Test_Data['train'][:5]['image'],Test_Data['train'][:5]['label']
```

```python
from transformers import pipeline
checkpoint = "openai/clip-vit-large-patch14"
detector = pipeline(model=checkpoint, task="zero-shot-image-classification")
```

```python
def find_predict(predictions):
  import numpy as np
  return np.array([item['score']for item in predictions]).argmax()

predictions = detector(x_test, candidate_labels=Test_Data["train"].features["label"].names)
```

```python
from sklearn.metrics import f1_score
y_pred=[find_predict(item) for item in predictions]
f1_score(y_test, y_pred, average='micro')
```
