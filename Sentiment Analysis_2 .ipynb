### Reading Data

```python
df=pd.read_csv(path+'train_data.csv')
df
```

### Similar Words for Guarantee and Warranty

```python
guarantee_warranty_similars=[
  'guarantee',
  'garantee',
  'guarentee'
  'guarantie',
  'guarntee',
  'guarentee',
  'warranty',
  'waranty',
  'warrenty',
  'warrantee',
  'warentee',
  'warrnaty',
]
```

### Number of Reviews that contain the Words

```python
df.reviewText.str.contains('|'.join(guarantee_warranty_similars)).sum()
```

```python
filtered_df=df[df.reviewText.str.contains('|'.join(guarantee_warranty_similars))]
filtered_df
```

### Mean-Score of Filtered Data

```python
filtered_df.groupby('reviewerID').agg({'overall':'mean'}).sort_values(by='overall',ascending=False)
```
